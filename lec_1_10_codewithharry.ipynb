# Summary Notes: Python Modules and Pip

This session covers the basics of **Modules**, the two main types of modules in Python, and how to use **pip** as a package manager.

---

## 1. What are Modules?

Modules are pre-written code files that you can "borrow" and use in your own programs. Instead of writing complex logic from scratch (like advanced mathematical formulas or machine learning algorithms), you can use modules written by experts.

**Key Benefits:**

* **Saves Time:** You don't have to reinvent the wheel.
* **Reduces Errors:** Code in modules is well-tested and used by millions.
* **Focus on Big Projects:** Allows you to focus on the unique parts of your project rather than basic utility code.

---

## 2. Types of Modules

Python modules are categorized into two types:

| Type | Description | Analogy |
| --- | --- | --- |
| **Built-in Modules** | Pre-installed with Python. You don't need to download anything extra. | Household items like plates or spoons that are already in your kitchen. |
| **External Modules** | Not included by default. These are written by other people and must be installed via `pip`. | Groceries like milk or bread that you have to go out and buy when needed. |

**Examples:**

* **Built-in:** `hashlib`, `os`, `random`.
* **External:** `pandas` (Data analysis), `scikit-learn` (Machine Learning), `tensorflow`.

---

## 3. What is Pip?

**Pip** is the package manager for Python. It acts as a "delivery service" that fetches external modules from the internet and installs them on your system.

* **Windows Command:** `pip install <module_name>`
* **Mac/Linux Command:** `pip3 install <module_name>`

> **Example:** To install the Pandas library, you would type `pip install pandas` in your terminal.

---

## 4. The REPL

**REPL** stands for **Read, Evaluate, Print, Loop**.
It is an interactive shell where you can type Python code and see the results immediately. You access it by simply typing `python` (Windows) or `python3` (Mac/Linux) in your terminal.

While useful for quick testing, most development is done by writing scripts in `.py` files.

---

## Key Takeaways

* Use the `import` keyword to bring a module into your code.
* Always try to use the latest version of Python (currently 3.x), as the basics remain consistent.

This lesson explores how to write non-executing code, handle special characters within strings, and master the advanced features of the `print()` function.

---

## 1. Python Comments

Comments are lines of text that the Python interpreter **ignores**. They are used to explain code logic to yourself or other developers.

### Types of Comments

* **Single-line Comment:** Created using the pound/hash (`#`) symbol.
* **Multi-line Comment:** Created by enclosing text in triple quotes (`'''` or `"""`).
> *Note: While technically these are "string literals" ignored by Python when not assigned to a variable, they are commonly used as multi-line comments.*



### Helpful Shortcuts

* **Comment/Uncomment Block:** Select the lines and press `Ctrl + /` (Windows/Linux) or `Command + /` (Mac).

---

## 2. Escape Sequence Characters

To insert characters that are "illegal" or have special meaning in a string, we use a **backslash (`\`)** followed by the character.

| Escape Sequence | Description | Result |
| --- | --- | --- |
| `\n` | **New Line** | Moves the text following it to a new line. |
| `\"` | **Double Quote** | Inserts a `"` inside a string already wrapped in double quotes. |
| `\'` | **Single Quote** | Inserts a `'` inside a string already wrapped in single quotes. |

---

## 3. Advanced Print Statement

The `print()` function is more powerful than just showing text; it can take several **optional parameters**:

### A. The `sep` (Separator) Parameter

By default, when you print multiple objects separated by a comma, Python adds a space between them. Use `sep` to change this.

* **Example:** `print("Hello", "World", sep="-")` → `Hello-World`

### B. The `end` Parameter

By default, `print()` adds a new line (`\n`) at the end of its output. Use `end` to specify what happens at the very end of the statement.

* **Example:** `print("Hello", end="009")` will print "Hello009" and keep the cursor on the same line for the next print statement.

---

### Key Takeaways

* **Comments** make your code readable and maintainable over time.
* **Escape sequences** allow you to format text output (like new lines) precisely.

# Summary Notes: Variables and Data Types

Welcome to Day 6! Today’s session focuses on how Python stores information in memory and the different "flavors" of data we can use.

---

## 1. What is a Variable?

A **Variable** is a container for storing data values.

* **Analogy:** Think of a kitchen container. It can hold sugar today, but you can empty it and put salt in it tomorrow.
* **How it works:** When you create a variable (e.g., `a = 1`), Python allocates a space in your **RAM** (memory) to store that value and labels it with the name `a`.

---

## 2. Data Types

A **Data Type** specifies the category of data a variable holds. This is crucial because Python needs to know what operations it can perform on that data (e.g., you can add two numbers, but you cannot add a number to a name).

### Common Built-in Data Types:

| Category | Type | Description | Example |
| --- | --- | --- | --- |
| **Numeric** | `int`, `float`, `complex` | Whole numbers, decimals, or complex math numbers. | `5`, `3.14`, `8 + 2j` |
| **Text** | `str` (String) | A sequence of characters wrapped in `" "` or `' '`. | `"Harry"` |
| **Boolean** | `bool` | Represents logical values: Yes or No. | `True`, `False` |
| **None** | `NoneType` | Represents the absence of a value. | `None` |

---

## 3. Sequence & Mapped Data

Sometimes we need to store a collection of items rather than just one.

### List vs. Tuple

* **List:** A collection of items that is **Mutable** (can be changed).
* **Tuple:** A collection of items that is **Immutable** (cannot be changed).
> **Memory Trick:** "Mutation" means change (like evolution). If it's *mutable*, you can mutate/change it!



### Dictionary (`dict`)

A collection of **key-value pairs**, similar to a real-world dictionary or a student report card.

* **Example:** `{"name": "Sakshi", "age": 20}`. Here, "name" is the key, and "Sakshi" is the value.

---

## 4. Key Functions & Concepts

* **`type()` function:** Use this to find out the data type of any variable.
* *Example:* `print(type(a))` might output `<class 'int'>`.


* **Everything is an Object:** In Python, every piece of data—whether a number or a list—is technically an "Object" belonging to a "Class."

---

### Important Rule: Same-Type Operations

You generally cannot perform operations between incompatible types.

* `print(5 + 5)` ✅ (Result: 10)
* `print("5" + "5")` ✅ (Result: "55" - this is called **concatenation**)

# Summary Notes: Type Casting in Python

Day #9 is all about **Type Casting** (also known as Type Conversion). This is the process of converting one data type into another to perform specific operations or avoid errors.

---

## 1. What is Type Casting?

Type casting allows you to change a variable from one type (e.g., a **String**) to another (e.g., an **Integer**).

**The Problem:**
If you have `a = "1"` and `b = "2"`, Python treats them as text.

* `print(a + b)` results in **"12"** (Concatenation).

**The Solution:**

* `print(int(a) + int(b))` results in **3**.

---

## 2. Explicit Type Casting

This is done **manually** by the programmer using built-in functions. You are telling Python exactly what to do.

* **Common Functions:**
* `int()`: Converts to an Integer.
* `float()`: Converts to a Floating-point number.
* `str()`: Converts to a String.
* `hex()`, `oct()`: Converts to Hexadecimal or Octal.



> **Warning:** The content must be valid for the conversion. You can convert `"27"` to an integer, but you cannot convert `"Harry"` to an integer—this will cause a **ValueError**.

---

## 3. Implicit Type Casting

This is done **automatically** by the Python interpreter to prevent data loss. Python converts a "lower" data type (like an Integer) to a "higher" data type (like a Float) during calculations.

**Example:**

```python
c = 1.9  # Float
d = 8    # Integer
print(c + d) # Result: 9.9 (Float)

```

In this case, Python automatically promoted the integer `8` to the float `8.0` so it could add it to `1.9` without losing the decimal information.

---

## 4. Summary Table

| Feature | Explicit Type Casting | Implicit Type Casting |
| --- | --- | --- |
| **Who does it?** | The Programmer | The Python Interpreter |
| **Methods used** | `int()`, `float()`, `str()`, etc. | Automatic (no method needed) |
| **Purpose** | To force a specific data type | To avoid data loss during math |

---

### Key Takeaway

Always be mindful of your data types! If your math looks like `"1" + "2" = "12"`, you likely need to use **Explicit Type Casting** to turn those strings into numbers.

# Summary Notes: Taking User Input in Python

Today, we take our programs from static to interactive! Input is the reason games and apps are addictive—they respond to **you**.

---

## 1. The `input()` Function

In Python, we use the `input()` function to get data from the user.

* **How it works:** When Python reaches an `input()` line, it pauses the program and waits for the user to type something and press **Enter**.
* **Variable Storage:** You must assign the input to a variable to save what the user typed.

**Example:**

```python
name = input("Enter your name: ")
print("Hello", name)

```

---

## 2. The "Greatest Doubt": String vs. Number

This is where most beginners get stuck. **The `input()` function always returns a String.**

### The Problem:

If you ask for two numbers and try to add them, Python treats them like text.

* User enters: `12` and `100`
* `print(x + y)` → Result: `"12100"`
* This is called **String Concatenation** (sticking text together side-by-side).

### The Solution:

If you want to do math, you must **Type Caste** the input into an integer or float immediately.

**Correct Way:**

```python
x = input("Enter first number: ")
y = input("Enter second number: ")
print(int(x) + int(y)) # Converts strings to integers before adding

```

---

## 3. Handling Errors (The "Value Error")

Python isn't a magician. If you tell Python to convert a word like `"Harry"` into an integer (`int("Harry")`), the program will crash.

* **Error Message:** `ValueError: invalid literal for int()`
* **Meaning:** Python is saying, "I can convert '12' to a number, but I have no idea how to turn 'Harry' into a number!"
