# Summary Notes: Function Arguments in Python üêç

In today's lesson, we dive deep into how to pass data into functions. There are four main ways to handle arguments, giving you a lot of flexibility in how your code works.

---

## 1. Types of Arguments

### A. Default Arguments

You can provide a default value when defining the function. If the user doesn't provide a value during the function call, the default is used.

```python
def average(a=9, b=1):
    print("The average is:", (a + b) / 2)

average(5) # a=5, b uses default (1). Output: 3.0
average()  # Uses defaults a=9, b=1. Output: 5.0

```

### B. Keyword Arguments

You can pass arguments using the `key = value` syntax. This allows you to ignore the order of parameters.

```python
average(b=9, a=21) # Order doesn't matter!

```

### C. Required Arguments

If an argument does **not** have a default value, it is mandatory. You must pass it, or Python will throw an error.

```python
def average(a, b=1): # 'a' is required
    pass

average(b=2) # ERROR: missing 'a'

```

### D. Variable-length Arguments

Sometimes you don't know how many numbers someone will pass.

* **Arbitrary Arguments (`*args`):** Python receives a **Tuple** of arguments.
* **Keyword Arbitrary Arguments (`**kwargs`):** Python receives a **Dictionary** of arguments.

```python
def average(*numbers): # numbers is a Tuple
    sum = 0
    for i in numbers:
        sum = sum + i
    return sum / len(numbers)

print(average(5, 6, 7, 1)) # Can pass as many as you want!

```

---

## 2. The `return` Statement

The `return` statement is used to send a value back to the "calling function."

* **Think of it as:** "Take this result and go back to where you came from."
* If a function doesn't have a `return` statement, it returns `None` by default.

```python
def average(*numbers):
    # logic...
    return 7 # The function hands over '7' to the caller

c = average(5, 6) # c now equals 7

```

---

## 3. Quick Reference Table

| Argument Type | Syntax Example | Key Feature |
| --- | --- | --- |
| **Default** | `def func(a=1):` | Uses default if value is missing. |
| **Keyword** | `func(b=2, a=1)` | Order of arguments doesn't matter. |
| **Required** | `def func(a):` | Must be provided by the caller. |
| **Variable (*)** | `def func(*args):` | Accepts multiple values as a **Tuple**. |
| **Variable (**)** | `def func(**kwargs):` | Accepts "key=value" pairs as a **Dictionary**. |

# Summary Notes: Python Lists üìù

Lists are one of the most powerful and versatile data types in Python. They allow you to store multiple items (even of different types) in a single variable.

---

## 1. What is a List?

A list is an **ordered collection** of data items.

* They are enclosed in square brackets `[]`.
* Items are separated by commas.
* **Important:** Lists are **mutable** (changeable). You can add, remove, or change items after the list is created.

```python
marks = [3, 5, 6, "Harry", True]
print(marks) # Output: [3, 5, 6, 'Harry', True]

```

---

## 2. Accessing Items (Indexing)

Every item in a list has an **index** starting from `0`.

| Index | 0 | 1 | 2 | 3 | 4 |
| --- | --- | --- | --- | --- | --- |
| **Value** | 3 | 5 | 6 | "Harry" | True |

* `marks[0]` is `3`.
* `marks[4]` is `True`.
* Accessing an index that doesn't exist (like `marks[10]`) results in an `IndexError`.

### Negative Indexing

Negative indexes count from the end.

* `marks[-1]` is the last element.
* **Pro-Tip:** To understand `marks[-3]`, convert it to positive: `len(marks) - 3`. For a list of length 5, `5 - 3 = 2`. So `marks[-3]` is the same as `marks[2]`.

---

## 3. List Slicing & Jump Index

You can extract a range of items from a list using the colon `:` syntax.
**Syntax:** `listName[start : end : jump]`

* **`marks[1:4]`**: Starts at index 1 and goes up to (but *not* including) index 4.
* **`marks[1:]`**: Starts at 1 and goes to the very end.
* **`marks[:4]`**: Starts at 0 and goes up to index 4.
* **Jump Index**: `marks[1:8:2]` will take every 2nd element within the range.

---

## 4. Checking for Elements

Use the `in` keyword to check if an item exists in a list. This also works for strings!

```python
if "Harry" in marks:
    print("Yes")
else:
    print("No")

```

---

## 5. List Comprehension üöÄ

List comprehension is a concise way to create new lists based on existing sequences or ranges.

**Syntax:** `[expression for item in iterable if condition]`

```python
# Create a list of squares of even numbers from 0 to 9
lst = [i*i for i in range(10) if i%2 == 0]
print(lst) # Output: [0, 4, 16, 36, 64]

```

---

### üí° Quick Summary Table

| Property | Description |
| --- | --- |
| **Mutable** | You can change the items inside. |
| **Ordered** | The items stay in the order you put them. |
| **Heterogeneous** | Can store integers, strings, booleans, etc., all in one list. |
| **Index starts at 0** | The first item is always at index 0. |

# Summary Notes: Python List Methods üõ†Ô∏è

In Python, **Methods** are built-in functions that you can use specifically on list objects to modify them, search through them, or organize them.

---

## 1. Modifying the List (In-place)

These methods change the original list directly. They do **not** create a new list.

* **`.append()`**: Adds an element to the very end of the list.
* **`.insert(index, value)`**: Adds an element at a specific position.
* **`.extend(list2)`**: Adds all elements of another list to the end of the current list.

```python
l = [11, 45, 1, 2, 4, 6]
l.append(7)           # l becomes [11, 45, 1, 2, 4, 6, 7]
l.insert(1, 899)      # Inserts 899 at index 1
l.extend([900, 1000]) # Adds multiple items at the end

```

---

## 2. Organizing and Reversing

* **`.sort()`**: Sorts the list in ascending order. Use `reverse=True` for descending order.
* **`.reverse()`**: Simply flips the current order of the list.

```python
l.sort()              # Sorts smallest to largest
l.sort(reverse=True)  # Sorts largest to smallest
l.reverse()           # Reverses current order

```

---

## 3. Searching and Counting

* **`.index(value)`**: Returns the index of the **first** occurrence of the value.
* **`.count(value)`**: Returns how many times the value appears in the list.

```python
print(l.index(1)) # Returns the position of '1'
print(l.count(1)) # Returns how many '1's are in the list

```

---

## 4. The Important `.copy()` Method ‚ö†Ô∏è

When you write `m = l`, you are **not** creating a new list. You are just giving the same list a second name (a reference). If you change `m`, `l` changes too. To create a completely independent duplicate, use `.copy()`.

```python
# The WRONG way
m = l 
m[0] = 0 # This changes l[0] too!

# The RIGHT way
m = l.copy()
m[0] = 0 # Now only m is changed; l stays original

```

---

## 5. List Concatenation

If you want to merge two lists into a **new** third list without changing the originals, use the `+` operator.

```python
l = [1, 2, 3]
m = [4, 5, 6]
k = l + m # k is [1, 2, 3, 4, 5, 6]

```

---

| **`insert()`** | Adds one item at a specific index. |
| **`extend()`** | Merges another list into the end. |
| **`sort()`** | Arranges items in order. |
| **`reverse()`** | Flips the order of items. |
| **`copy()`** | Creates a separate duplicate. |

# Summary Notes: Python Tuples üéí

A **Tuple** is a data type in Python very similar to a List, but with one critical difference: **Tuples are immutable.** This means once a tuple is created, it cannot be changed, modified, or altered.

---

## 1. What is a Tuple?

* **Syntax:** Defined using round brackets `()`.
* **Ordered:** The items have a defined order that will not change.
* **Immutable:** You cannot add, remove, or change items after creation.

```python
tup = (1, 2, 76, 342, 32, "Harry", True)
print(type(tup)) # Output: <class 'tuple'>

```

### The Single Element Trap ‚ö†Ô∏è

If you want to create a tuple with only **one** element, you **must** include a comma after the item. Otherwise, Python will think it's just a regular integer or string in parentheses.

* `tup = (1)`  Python sees this as an **int**.
* `tup = (1,)`  Python sees this as a **tuple**.

---

## 2. Accessing Tuple Items

Indexing and slicing work exactly like they do in Lists and Strings.

### Positive and Negative Indexing

* **Positive:** Starts from `0` (left to right).
* **Negative:** Starts from `-1` (right to left).

**How to handle Negative Indexing:**
Always convert it to positive indexing to avoid confusion:

For a tuple of length 7, `tup[-1]` is the same as `tup[7 - 1]` which is `tup[6]`.

---

## 3. Tuple Slicing

You can extract a portion of a tuple. This does **not** change the original tuple; instead, it returns a **new** tuple.

```python
tup2 = tup[1:4] # Elements from index 1 to 3
print(tup2)     # Output: (2, 76, 342)

```

---

## 4. Key Differences: Tuple vs. List

| Feature | List | Tuple |
| --- | --- | --- |
| **Brackets** | Square `[]` | Round `()` |
| **Mutability** | **Mutable** (Can be changed) | **Immutable** (Cannot be changed) |
| **Methods** | Many (append, sort, etc.) | Very few (count, index) |
| **Use Case** | When data needs to change. | For constants or data that must stay safe. |

---

## 5. Checking for Elements

Use the `in` keyword to see if an item exists within the tuple. This returns `True` or `False`.

```python
if 342 in tup:
    print("Yes, it is present!")

```

---

### üí° Summary Tip

Think of a **List** like a **whiteboard** (you can write and erase) and a **Tuple** like a **stone carving** (once it's there, it's permanent).

# Summary Notes: Python f-Strings (Fast Strings) ‚ö°

Introduced in Python 3.6, **f-strings** (formatted string literals) are the most modern, efficient, and readable way to format strings. They allow you to embed expressions and variables directly inside string literals using curly braces `{}`.

---

## 1. The Evolution of String Formatting

Before f-strings, Python used older methods that were often harder to read or more verbose.

### A. The `.format()` Method (Older approach)

You define placeholders with `{}` and pass variables into the `.format()` method.

```python
name = "Harry"
country = "India"
letter = "Hey my name is {} and I am from {}"
print(letter.format(name, country))

```

* **The Problem:** If you have many variables, it becomes difficult to track which variable goes into which bracket. You have to manage indexes like `{0}` and `{1}` manually.

### B. The f-String Approach (Modern approach)

Simply prefix your string with `f` or `F` and put the variable name directly inside the `{}`.

```python
print(f"Hey my name is {name} and I am from {country}")

```

* **The Benefit:** It is much more concise and easier to debug because you can see the variable names exactly where they will appear in the text.

---

## 2. Advanced f-String Features

### Formatting Numbers (Decimal Places)

You can control how numbers are displayed (e.g., rounding) using a colon `:` followed by a format specifier.

* **`.2f`**: Rounds the number to 2 decimal places.

```python
price = 49.09999
txt = f"For only {price:.2f} dollars!" 
print(txt) # Output: For only 49.10 dollars!

```

### Performing Expressions

f-strings can evaluate math or logic directly inside the braces.

```python
print(f"{2 * 30}") # Output: 60

```

---

## 3. How to Print Literal Curly Braces? üí°

Sometimes you want to show the curly braces in your output (like when writing a tutorial) rather than replacing them with a variable. To do this, use **double curly braces**.

| Goal | Syntax | Output |
| --- | --- | --- |
| **Variable Value** | `f"Value: {name}"` | `Value: Harry` |
| **Literal Braces** | `f"Value: {{name}}"` | `Value: {name}` |

---

## 4. Key Takeaways Table

| Feature | Syntax Example | Use Case |
| --- | --- | --- |
| **Variable Insertion** | `f"Hello {name}"` | General string building. |
| **Float Formatting** | `{val:.3f}` | Limiting decimal points (e.g., 3.142). |
| **Escaping Braces** | `{{text}}` | Printing raw braces in an f-string. |
| **Evaluation** | `{price * 2}` | Quick calculations inside text. |

---

### üí° Pro-Tip

f-strings are faster than both `%` formatting and `.format()` because they are evaluated at runtime rather than being constant strings that require a function call.

# Summary Notes: Docstrings and PEP-8 üìú

In this session, we covered two vital topics that often appear in Python interviews: **Docstrings**, which help document your code, and **PEP-8**, the official style guide for writing clean Python.

---

## 1. What are Docstrings?

**Docstrings** (Documentation Strings) are string literals that appear right after the definition of a function, method, class, or module.

### How they differ from Comments:

* **Comments:** Ignored by the Python interpreter. They are for humans reading the source code.
* **Docstrings:** Not ignored. They are stored in the `__doc__` attribute of the object and can be accessed at runtime.

### Example:

```python
def square(n):
    '''Takes in a number n, returns the square of n'''
    return n**2

print(square.__doc__)

```

**Output:** `Takes in a number n, returns the square of n`

### Important Rule ‚ö†Ô∏è

The Docstring **must** be the very first statement in the function body. If you put any other code (even a simple `print`) before it, Python will no longer recognize it as a Docstring, and `__doc__` will return `None`.

---

## 2. PEP-8 (Python Enhancement Proposal)

**PEP-8** is a document that provides guidelines and best practices on how to write Python code. It was created in 2001 to improve the **readability** and **consistency** of Python code.

* **Readability counts:** Code is read much more often than it is written.
* **Consistency:** PEP-8 ensures that Python code looks similar regardless of who wrote it.

---

## 3. The Zen of Python (The Easter Egg) ü•ö

Python has a built-in "poem" or philosophy called **The Zen of Python**. It highlights the guiding principles of the language.

To see it, simply run this in your Python shell:

```python

```

### Key Principles from the Poem:

* **Beautiful is better than ugly.**
* **Simple is better than complex.**
* **Readability counts.**
* **Flat is better than nested.**

As you grow as a programmer and complete 1-2 end-to-end projects, these lines will start to make much more sense to you!

# Summary Notes: Recursion in Python üîÑ

**Recursion** is a programming technique where a function calls itself in order to solve a problem. It is essentially "defining something in terms of itself."

---

## 1. How Recursion Works

A recursive function generally has two main parts:

1. **Base Case:** The condition under which the function stops calling itself. Without this, the function would run forever (causing a stack overflow).
2. **Recursive Case:** The part where the function calls itself with a modified argument, moving closer to the base case.

---

def factorial(n):
    # Base Case: 0! or 1! is 1
    if (n == 0 or n == 1):
        return 1
    # Recursive Case
    else:
        return n * factorial(n - 1)

## 5. Comparison: Recursion vs. Iteration (Loops)

| Feature | Recursion | Iteration (Loops) |
| --- | --- | --- |
| **Definition** | Function calls itself. | Repeated execution of a code block. |
