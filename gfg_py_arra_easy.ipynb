Array Search


Given an array, arr[] of n integers, and an integer element x, find whether element x is present in the array. Return the index of the first occurrence of x in the array, or -1 if it doesn't exist.

Examples:

Input: arr[] = [1, 2, 3, 4], x = 3
Output: 2
Explanation: For array [1, 2, 3, 4], the element to be searched is 3. Since 3 is present at index 2, the output is 2.


```python
class Solution:
    def search(self, arr, x):
        # code here
        for a in range (0, len(arr)):
            if arr[a] ==x:
                return a
        else:
            return -1
                


c=Solution()
c.search([1,2,3,4],3)
```

This is called Linear Search.

Time Complexity: O(n)
Space Complexity: O(1)

DSA approach 

If the array is sorted, a DSA approach would be Binary Search

Time Complexity: O(log n)

Key point first (very important)

ðŸ‘‰ Binary Search only works on a sorted array
Your array:

arr = [10, 3, 4, 1, 2, 16]


is NOT sorted, so we cannot directly apply Binary Search.

So the DSA approach here becomes a 2-step process.

```python
class Solution:
    def search(self, arr, x):
        # code here
        arr.sort()
        low =0
     
        high=len(arr)-1
        while low<=high:
            mid = (low+high)//2
            if arr[mid]==x:
                return mid
            elif arr[mid]<x:
                low =low +1
            else:
                high =mid -1
        return -1


c = Solution()
c.search([10,3,4,1,2,16], 4)
                

            



        
       
                
```
