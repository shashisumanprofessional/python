## 1. Creating Strings

In Python, a string is created by enclosing characters in quotes. You have three main ways to do this:

* **Single Quotes (`' '`):** `friend = 'Rohan'`

> **Pro Tip:** Use triple quotes if you are writing a poem, a long paragraph, or even a piece of text that contains both single and double quotes internally.



---

## 2. String Indexing

A string is like a sequence of characters. Each character has a specific "address" called an **Index**.

* **Counting starts at 0.** The first character is `[0]`, the second is `[1]`, and so on.
* **Index Error:** If you try to access an index that doesn't exist (e.g., trying to get the 6th character of a 5-letter word), Python will throw an `IndexError`.

---

## 3. Looping Through a String

If you want to access every character in a string one by one, you can use a **`for` loop**. This is much more efficient than printing each index manually.

**Example:**

```python
name = "Harry"
for character in name:
    print(character)

```

**Output:**

```text
H
a
r
r
y

```

---

## 4. Why Use Different Quotes?

If your string contains a quote, you can avoid errors by wrapping it in the *other* type of quote.

* **Wrong:** `sentence = "He said, "I am good""` (This confuses Python).
* **Right:** `sentence = 'He said, "I am good"'` (Wrapping double quotes inside single quotes).
* **Right:** `sentence = "He said, 'I am good'"` (Wrapping single quotes inside double quotes).

---

### Key Takeaway

Strings are **sequences**. You can access their parts using indices, write them across multiple lines using triple quotes, and process them character-by-character using loops.

# Summary Notes: String Slicing and Length

Welcome back! Today, we are learning how to "cut" or **slice** strings to extract exactly the parts we need. This is one of the most powerful tools when working with text data.

---

## 1. Finding String Length: `len()`

Before you can slice a string properly, you need to know how long it is. We use the `len()` function for this.

* **Example:** `name = "Harry"`
* `print(len(name))` outputs `5`.

---

## 2. Basic String Slicing

Slicing allows you to get a range of characters from a string using square brackets `[start:end]`.

**The Golden Rule:** > Python includes the **start** index but **excludes** the **end** index.

> It goes from `start` to `end - 1`.

highlighting "MAN"]

| Code | Output | Explanation |
| --- | --- | --- |
| `fruit = "Mango"` |  | Our base string. |
| `fruit[0:4]` | `Mang` | Index 0, 1, 2, and 3. |
| `fruit[:4]` | `Mang` | Python assumes 0 if the start is blank. |
| `fruit[1:4]` | `ang` | Starts at index 1, stops before 4. |
| `fruit[2:]` | `ngo` | Starts at index 2 and goes to the very end. |
| `fruit[:]` | `Mango` | Copies the whole string. |

---

## 3. Negative Slicing

Negative slicing can be confusing at first, but there is a simple trick to solve it: **Always add the length of the string to the negative numbers.**

**Formula:** `fruit[-a:-b]` is interpreted by Python as `fruit[len(fruit)-a : len(fruit)-b]`

**Example with "Mango" (Length = 5):**

* `fruit[-3:-1]`
* Calculation: `5 - 3 = 2` and `5 - 1 = 4`
* Result: `fruit[2:4]` â†’ **"ng"**

---

## 4. Quick Quiz Challenge

Think you've mastered it? Try to solve this without running the code:

```python
nm = "Harry"
# What is the output of the line below?
print(nm[-4:-2])

```

### ðŸ§  How to solve it:

1. Find the length of `nm`.
2. Add that length to `-4` (this is your start index).
3. Add that length to `-2` (this is your end index).
4. Remember: Include the start, exclude the end!

---

### Key Takeaway

String slicing is all about indices. Use `[start:end]` to get your substring, and don't forget that the character at the `end` index is never included!

---

## 1. The Core Rule: Strings are Immutable

In Python, **Strings are immutable**, meaning you cannot change a string once it is created.

* When you use a method like `upper()` or `replace()`, Python **does not** change the original variable.
* Instead, it creates and returns a **new string**.

---

## 2. Essential String Methods

Here are the most commonly used methods to transform and inspect text:

### Text Transformation

| Method | Description | Example |
| --- | --- | --- |
| **`upper()`** | Converts all characters to uppercase. | `"hi".upper()`  `"HI"` |
| **`lower()`** | Converts all characters to lowercase. | `"HI".lower()`  `"hi"` |
| **`strip()`** | Removes leading and trailing whitespaces. | `"  hey  ".strip()`  `"hey"` |
| **`rstrip()`** | Removes trailing characters (default: space). | `"hi!!!" .rstrip("!")`  `"hi"` |
| **`replace(old, new)`** | Replaces all occurrences of a substring. | `"apple".replace("p", "b")`  `"abble"` |
| **`capitalize()`** | Capitalizes the first letter and lowers the rest. | `"hELLO".capitalize()`  `"Hello"` |
| **`title()`** | Capitalizes the first letter of every word. | `"hello world".title()`  `"Hello World"` |
| **`swapcase()`** | Swaps upper to lower and vice-versa. | `"PyThOn".swapcase()`  `"pYtHoN"` |

### Text Inspection (Returns True/False)

| Method | Description |
| --- | --- |
| **`endswith()`** | Checks if the string ends with a specific value. |
| **`startswith()`** | Checks if the string starts with a specific value. |
| **`isalnum()`** | Returns `True` if the string contains only letters and numbers (no spaces/symbols). |
| **`isalpha()`** | Returns `True` if the string contains only letters. |
| **`islower()` / `isupper()**` | Checks the casing of the whole string. |
| **`isprintable()`** | Returns `False` if there are characters like `\n` (new line) or `\t` (tab). |
| **`isspace()`** | Returns `True` only if the string is made of empty spaces. |

---

## 3. Finding & Counting

* **`count("substring")`**: Returns how many times a word or character appears.
* **`find("substring")`**: Returns the index of the first occurrence. Returns **-1** if not found.
* **`index("substring")`**: Same as find, but **crashes the program** with an error if the substring is not found. Use this if you are 100% sure the item exists.

---

## 4. The `split()` Method

This is a powerful method that turns a string into a **List**.

* **Example:** `"Harry is good".split(" ")`
* **Result:** `['Harry', 'is', 'good']`
It "cuts" the string at every space (or whatever separator you choose).

# Summary Notes: If/Else Conditional Statements

Welcome to Day #14! Today we learn how to give our Python programs "brains" so they can make decisions. Just like you decide to skip school if it rains, your code can now decide what to do based on user input.

---

## 1. The `if-else` Logic

Python uses `if` and `else` to execute different blocks of code depending on whether a condition is **True** or **False**.

**The Structure:**

* **`if`**: The starting condition.
* **`elif`** (Short for "else if"): Use this if you have multiple specific conditions to check.
* **`else`**: The "catch-all" that runs if none of the above conditions are met.

---

## 2. Conditional Operators

To make decisions, Python compares values using these operators. They always return a **Boolean** (`True` or `False`).

| Operator | Meaning | Example |
| --- | --- | --- |
| **`>`** | Greater than | `age > 18` |
| **`<`** | Less than | `age < 18` |
| **`>=`** | Greater than or equal to | `age >= 18` |
| **`<=`** | Less than or equal to | `age <= 18` |
| **`==`** | Equal to | `age == 18` |
| **`!=`** | Not equal to | `age != 18` |

> **Important:** Never confuse `=` (assignment) with `==` (comparison). Use `==` when you want to check if two things are the same!

---

## 3. Indentation: The Rule of Space

In languages like C++, we use curly brackets `{ }`. In Python, we use **Indentation** (spaces).

* A **colon (`:`)** marks the start of a block.
* Every line inside that block **must be indented** (usually 4 spaces or 1 tab).
* If you stop indenting, you have "exited" the block.

```python
if (age > 18):
    print("You can drive")  # Indented: inside the 'if'
print("This always prints") # Not indented: independent of 'if'

```

---

## 4. Nested `if` Statements

You can put an `if` statement inside another `if` statement. This is called **Nesting**. It allows for very complex decision-making.

**Example:**

```python
num = 15
if (num > 0):
    print("Number is positive")
    if (num <= 10):
        print("Number is between 1-10")
    elif (num > 10 and num <= 20):
        print("Number is between 11-20")

```

---

## 5. Summary of the `if-elif-else` Ladder

1. Python checks the `if` condition. If it's True, it runs the code and **skips** the rest of the ladder.
2. If the `if` is False, it checks each `elif` one by one.
3. If **nothing** is True, the `else` block runs.

### Key Takeaway

Conditional statements are the foundation of logic in programming. Master the **indentation** and **comparison operators**, and you can build anything from a simple calculator to complex game logic!

# Summary Notes: Match Case Statements in Python

Welcome to Day #16! Today we are discussing a relatively new feature introduced in **Python 3.10**: the **Match Case** statement.

If you have used languages like C, C++, or Java, you probably know this as the **Switch Case**. Itâ€™s a cleaner, more organized way to handle multiple conditions without writing endless `if-elif-else` blocks.

---

## 1. What is Match Case?

A Match Case statement takes a variable and compares its value against several patterns. When it finds a match, it executes the code block associated with that pattern.

**Basic Syntax:**

```python
x = int(input("Enter a value: "))

match x:
    case 0:
        print("x is zero")
    case 4:
        print("x is four")
    case _:
        print("This is the default case (x is something else)")

```

---

## 2. Key Differences from C/C++ (No "Break" Needed!)

One of the best things about Python's `match` statement is that you **don't need a `break` statement**.

* In C/C++, if you don't use `break`, the code "falls through" and executes the next cases too.
* In **Python**, only the **first** matching case is executed, and then the program automatically exits the match block.

---

## 3. The Default Case: Underscore (`_`)

The underscore `_` acts as a **wildcard**. It matches anything. You should place it at the very end of your match block to handle any values that weren't caught by the specific cases above. It's exactly like the `else` in an `if-else` ladder.

---

## 4. Advanced: Match Case with "If" (Guards)

You can add an `if` condition inside a `case` for more complex logic. This is known as a **Guard**.

**Example:**

```python
match x:
    case _ if x < 10:
        print("x is a small positive number")
    case _ if x > 100:
        print("x is a very large number")
    case _:
        print(x)

```

---

## 5. Important: Python Version Check

Match Case **only works in Python 3.10 and above**.

* If you are using an older version (like 3.8 or 3.9), your code will throw a `SyntaxError`.
* **In Replit:** You may need to select a "Python 3.10" template specifically if the default version is older.
* **In VS Code:** Just make sure your installed Python interpreter is updated.

---

Here is how `match` works across different data types:

---

### 1. Matching Strings

You can use it to handle text commands or user inputs. It is much cleaner than using multiple `if val == "text":` statements.

```python
command = "save"

match command:
    case "open":
        print("Opening file...")
    case "save" | "export":  # You can use | (OR) to match multiple values
        print("Saving data...")

```

---

### 2. Matching Lists (Sequence Patterns)

This is where `match` shines. You can check the length and the content of a list at the same time.

```python
data = ["load", "image.png", 50]

match data:
    case ["load", filename]:  # Matches a list with exactly 2 items
        print(f"Loading {filename}")
    case ["load", filename, priority]: # Matches 3 items

    case []:
        print("The list is empty")

```

---

### 3. Matching by Data Type

You can use `match` to check what **kind** of data was sent. This is very useful when a variable might be a string in one case and a number in another.

```python
def process_data(val):

            print(f"Processing integer: {n}")
        case str(s):
            print(f"Processing string: {s}")
        case float(f):
            print(f"Processing float: {f}")
        case _:
            print("Unsupported data type")

process_data(100)    # Matches int
process_data("Hi")   # Matches str

```

# Summary Notes: Introduction to Loops and `for` Loops

Imagine you have to fill 100 water bottles. You wouldn't write a separate instruction for each bottle; you would just say, "Repeat the filling process for every bottle." In programming, this is exactly what **Loops** do. They allow us to execute a block of code multiple times without rewriting it.

---

## 1. Why do we need Loops?

If you wanted to print numbers from 1 to 20,000, writing 20,000 `print()` statements would be impossible.

* **Efficiency:** Write the logic once, and let the computer repeat it.
* **Automation:** Perfect for processing items in a list, characters in a string, or lines in a file.

---

## 2. The `for` Loop

The `for` loop in Python is used to iterate over a **sequence** (like a String or a List). It goes through each item one by one.

### Iterating over a String:

```python
name = "Harry"
for char in name:
    print(char)
# Output: H, a, r, r, y (each on a new line)

```

### Iterating over a List:

```python
colors = ["Red", "Green", "Blue"]
for color in colors:
    print(color)
    for i in color:
        print(i) # This is a nested loop!

# Summary Notes: The `while` Loop in Python

While the `for` loop is great for iterating over a fixed sequence, the **`while` loop** is designed to repeat a block of code as long as a specific **condition** remains **True**. It is perfect for situations where you don't know exactly how many times you need to repeat the task.

---

## 1. How the `while` Loop Works

The loop checks the condition first. If it is `True`, the code inside runs. After the code finishes, it goes back to the top and checks the condition again. This repeats until the condition becomes `False`.

**Basic Syntax:**

```python
i = 0
while (i < 3):
    print(i)
    i = i + 1  # This is crucial!

```

---

## 2. The Importance of the "Counter"

Unlike a `for` loop, a `while` loop doesn't automatically move to the next item. You must manually update your variable (increment or decrement).

* **Incrementing:** `i = i + 1` (moving upwards)
* **Decrementing:** `count = count - 1` (counting down)

> **âš ï¸ Warning: The Infinite Loop**
> If your condition never becomes `False` (for example, if you forget to increment `i`), the loop will run forever! This can crash your program or slow down your computer.

---

## 3. Using `else` with `while`

Python allows you to use an `else` block with a `while` loop. The `else` block executes **once** the loop condition becomes `False`.

```python
count = 5
while (count > 0):
    print(count)
    count -= 1
else:
    print("I am inside else")

```

---

## 4. `while` vs `for` Loops

| Feature | `for` loop | `while` loop |
| --- | --- | --- |
| **Best Use** | Iterating over a sequence (list, string, range). | Repeating based on a condition (like user input). |
| **Control** | Automatic progression. | Manual progression (must update the variable). |
| **Complexity** | Simple syntax for counting. | Better for complex logic and dynamic conditions. |

---

## 5. Emulating the "Do-While" Loop

In some languages (like C++ or Java), a `do-while` loop executes the code **at least once** before checking the condition. Python doesn't have a built-in `do-while`, but we can emulate it by placing the input/code once before the loop starts:

```python
# Emulating a do-while loop
i = int(input("Enter number: "))
print(i) # Executes at least once
while (i <= 38):
    i = int(input("Enter number: "))
    print(i)

# Summary Notes: Break and Continue Statements

In this lesson, we cover how to control the flow of your loops more precisely. Sometimes you want to stop a loop entirely, and sometimes you just want to skip a specific part of it.

---

## 1. The `break` Statement

The `break` statement is like an "emergency exit." It completely **terminates** the loop it is in and moves the program to the next line of code outside that loop.

* **Job:** Exit the loop entirely.
* **Common Use:** Stop searching once you find what you're looking for.

```python
for i in range(12):
    if (i == 10):
        break  # Exit the loop when i is 10
    print(f"5 X {i+1} = {5*(i+1)}")

print("Loop ko chodkar nikal gaya (Left the loop)")

```

---

## 2. The `continue` Statement

The `continue` statement is like "skipping a turn." It doesn't stop the whole loop; it only stops the **current iteration** and jumps immediately to the next one.

* **Job:** Skip the rest of the code in the current loop body.
* **Common Use:** Skip specific items that don't meet your criteria.

```python
for i in range(12):
    if (i == 10):
        print("Skip the iteration")
        continue  # Skip everything below for this i, go to i=11
    print(f"5 X {i} = {5*i}")

```

---

## 3. Comparison: `break` vs `continue`

| Statement | Effect on Loop | What happens next? |
| --- | --- | --- |
| **`break`** | Stops the entire loop. | Code outside the loop runs. |
| **`continue`** | Stops only the current round. | The next round (iteration) starts. |

---

## 4. Emulating a "Do-While" Loop in Python

As discussed in Day #18, Python doesn't have a built-in `do-while` loop. However, we can use an **infinite while loop** and a **break statement** to create one.

In a `do-while` loop, the code runs **at least once**, and then we check the condition to see if we should continue.

**The Python way:**

```python
i = 0
while True:
    print(i)
    i = i + 1
    if not (i < 100): # The condition to keep going
        break        # If condition fails, exit

# Summary Notes: Introduction to Python Functions ðŸš€

In programming, a **Function** is like a "codeword" or a "recipe." Instead of writing the same 50 lines of code every time you need to do a specific task, you wrap that logic into a function and just call its name whenever you need it.

---

## 1. Why use Functions?

* **Dry (Don't Repeat Yourself):** You don't have to copy-paste the same logic multiple times.
* **Easy Maintenance:** If there is a mistake, you only fix it in **one** place (inside the function), and it updates everywhere the function is used.
* **Organization:** It breaks a big, scary program into smaller, manageable "entities" (like a function for "login," a function for "processing payment," etc.).

---

## 2. Defining and Calling a Function

To create a function, you use the `def` keyword.

### The Syntax:

```python
def function_name(parameter1, parameter2):
    # This is the function body
    # Logic goes here (indented)
    pass 

```

### Real Example:

Let's create a function to find the greater of two numbers:

```python
def isGreater(a, b):
    if (a > b):
        print("First number is greater")
    else:
        print("Second number is greater or equal")

# Calling the function
isGreater(9, 8)
isGreater(5, 12)

```

---

## 3. The `pass` Keyword

Sometimes you want to create the "shell" of a function but write the logic later. If you leave a function body empty, Python will throw an error. To prevent this, use the `pass` statement.

* **`pass`** tells Python: "I know there's no code here yet; just ignore it and move on."

```python
def calculateLesser(a, b):
    pass  # I will write this logic later!

```

---

## 4. Built-in vs. User-Defined Functions

Python gives you two types of functions:

1. **Built-in Functions:** Pre-defined functions that are always available.
* Examples: `print()`, `len()`, `min()`, `max()`, `sum()`, `type()`.


2. **User-Defined Functions:** Functions that **you** create using the `def` keyword to solve specific problems.

| **Parameters** | Variables inside the `()` that act as placeholders for data. |
| **Indentation** | The "Tab" or spaces that tell Python what code belongs to the function. |
| **Calling** | Using the function name with `()` to actually run the code. |
